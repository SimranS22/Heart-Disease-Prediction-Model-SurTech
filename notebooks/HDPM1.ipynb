### Step 8: Data Modelling

* Train-Test-Split 
* Fit the model to perform predictions
* Comparing the following algorithms to see which model performs the best

**Splitting the Dataset**

```python
X = df.drop(columns='Target',axis=1)
Y= df['Target'] 
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train , X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

**Encoding Text Data**

```python
trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False,drop='first'),['Sex','ChestPainType','RestingECG','ExerciseAngina','ST_Slope'])
]
,remainder='passthrough')
```

**Logistic Regression**

```python
pipelr = Pipeline(steps=[
    ('step1',trf),
    ('step2',LogisticRegression(solver='liblinear'))
])
```

```python
pipelr.fit(X_train,Y_train)
```

```python
Y_pred_lr = pipelr.predict(X_test)
```

**Random Forest Classifier**

```python
piperfc = Pipeline(steps=[
    ('step1',trf),
    ('step2',RandomForestClassifier(n_estimators=10))
])
```

```python
piperfc.fit(X_train,Y_train)
```

```python
Y_pred_rfc = piperfc.predict(X_test)
```

### Step 9: Model Evaluation

**Logistic Regression**

```python
accuracy_score(Y_test,Y_pred_lr)
```

**Random Forest Classifier**

```python
accuracy_score(Y_test,Y_pred_rfc)
```
